# Use case 1: Response surface

A quadratic function to study. The user provides this. It is usually a call to some simulation code---not a python function.

```python
import numpy as np
class MyQuadratic():
    def __init__(self):
        self.e = np.exp(np.array([2,-1,-2,-3,-4,-5,-6,-7,-8,-9.]))
        self.W = np.linalg.qr(np.random.normal(size=(10,10)))[0]
        self.A = np.dot(self.W,np.dot(np.diag(self.e),self.W.T))
    def __call__(self,x):
        return 0.5*np.dot(x,np.dot(self.A,x.T)),np.dot(self.A,x.T)

fun = MyQuadratic()
```

Discover the active subspace of the function.

```python
import active_subspaces as ac
M,m = 100,10
X = np.random.uniform(-1.,1.,size=(M,m))

# sample the function. often done offline. 'sample_function' is just for convenience
F,dF = ac.sample_function(X,fun,dflag=True)

# compute the active subspace components
k = 6
e,W,e_br,sub_br = ac.compute_active_subspace(dF,k,n_boot=1000)

# make some plots
ac.plot_eigenvalues(e,e_br=e_br)
ac.plot_subspace_errors(sub_br)
ac.plot_eigenvectors(W[:,:6])
y = np.dot(X,W[:,0])
ac.sufficient_summary_plot(y,F)
```

We exploit the one-dimensional active subspace to construct a response surface. 

```python
# Get a design on the active variable
n = 1 # dimension of the active subspace
N = [5] # number points in the active variable
NMC = 3 # number of monte carlo samples
X2,ind,Y = ac.response_surface_design(W,n,N,NMC,bflag=1)

# sample the function at the design points
F2 = ac.sample_function(X2,fun,dflag=True)[0]

# compute conditional expectations
Ny = Y.shape[0]
Fy = np.zeros((Ny,1))
Vy = np.zeros((Ny,1)) # empirical variance
for i in range(Ny):
    Fy[i] = np.sum(F2[ind==i])/np.sum(ind==i)

# build the response surface on the active variable
gp = ac.GaussianProcess(2)
gp.train(Y,Fy,e=e,gl=0.0,gu=10.0)

# plot the response surface
import matplotlib.pyplot as plt
y2 = np.linspace(np.amin(Y),np.amax(Y),101).reshape((101,1))
f2 = gp.predict(y2)[0]
plt.figure()
plt.plot(y,F,'bo')
plt.plot(y2,f2,'k-',Y,Fy,'ro')
plt.xlabel('Active variable')
plt.ylabel('Function')
plt.grid(True)
plt.show()
```
