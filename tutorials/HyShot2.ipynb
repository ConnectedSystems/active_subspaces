# Active Subspace Analysis Notebook for HyShot II Scramjet

**DESCRIPTION**

This data set comes from a numerical simulation of a hypersonic scramjet engine inspired by the HyShotII vehicle. The input variables characterize the inflow conditions of 3d compressible flow in the combustor. The outputs are the pressure at the combustor exit for two different fuel injection rates. 

**INPUTS (7)**

Variable | Lower bound | Upper bound | Density
--- | --- | --- | ---
Angle of attack | 2.6 | 4.6 | Uniform
Turbulence intensity | 0.1 | 0.9 | Uniform
Turbulence length scale | 122.5 | 367.5 | Uniform
Stagnation pressure | 1.6448e7 | 1.9012e7 | Uniform
Stagnation enthalpy | 3.0551e6 | 3.4280e6 | Uniform
Cowl transition location | 0.03 | 0.07 | Uniform
Ramp transition location | 0.087 | 0.203 | Uniform

**OUTPUTS (2)**

Variable | Description
--- | --- 
Exit press 0.30 | Pressure at the combustor exit for equivalence ratio (fuel injection) 0.30
Exit press 0.35 | Pressure at the combustor exit for equivalence ratio (fuel injection) 0.35

```python
import numpy as np
import pandas as pn
import active_subspaces as ac
import matplotlib.pyplot as plt
```

Import the data set, distinguish inputs (X) and outputs (F), and set the labels.

```python
df = pn.DataFrame.from_csv('../data/HyShot2.txt')
data = df.as_matrix()
X = data[:,0:7]
F = data[:,7:9]
M,m = X.shape
labels = df.keys()
in_labels = labels[:7]
out_labels = labels[7:9]
```

Normalize the inputs to the interval [-1,1].

```python
lb = np.array([2.6,0.1,122.5,1.6448e7,3.0551e6,0.6*0.05,0.6*0.145])
ub = np.array([4.6,1.9,367.5,1.9012e7,3.4280e6,1.4*0.05,1.4*0.145])
bn = ac.BoundedNormalizer(lb,ub)
XX = bn.normalize(X)
```

Choose the output of interest.

```python
out_index = 0 # either 0 or 1
f = F[:,out_index]
out_label = out_labels[out_index]
```

Check for a one-dimensional active subspace with the quick check.

```python
w = ac.quick_check(XX,f,in_labels=in_labels,out_label=out_label)
```

Construct a univariate response surface on the active variable.

```python
# evaluations of the active variable
Y = np.dot(XX,w).reshape((M,1))

# train the response surface on the active variable
pr = ac.PolynomialRegression(2)
pr.train(Y,f)

# plot the response surface of he active variable
y0 = np.dot(w.T,np.sign(w)) # need better code to do this
if y0 < -y0:
    yl,yu = y0,-y0
else:
    yl,yu = -y0,y0
yy = np.linspace(yl,yu,101).reshape((101,1))
gg,dgg,vgg = pr.predict(yy,compvar=True)

plt.figure()
plt.plot(Y,f,'bo')
plt.plot(yy,gg,'k-')
plt.plot(yy,gg+2*np.sqrt(vgg),'k--')
plt.plot(yy,gg-2*np.sqrt(vgg),'k--')
plt.xlabel('Active variable')
plt.ylabel(out_label)
plt.grid(True)
```

Use the response surface to estimate a density function of the output.

```python
# Sample the active subspace response surface
nevals = 10000
XXX = np.random.uniform(-1.0,1.0,size=(nevals,m))
YYY = np.dot(XXX,w).reshape((nevals,1))
ggg = pr.predict(YYY)[0]

# plot the histogram
from scipy.stats import gaussian_kde
kde = gaussian_kde(ggg)
z = np.linspace(0.9*np.amin(ggg),1.1*np.amax(ggg))

plt.plot(z,kde(z),'b-')
plt.xlabel(out_label)
plt.ylabel('Gaussian KDE')
plt.grid(True)
plt.show()
```
